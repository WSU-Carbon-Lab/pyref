# Catalog-only structure and naming QC

Uses only `catalog.db` (set `CATALOG` in the next cell or export `PYREF_CATALOG_DB` before starting Jupyter). No NAS path is required for metadata.

**Pixels:** Loading FITS requires files at stored `file_path` values; skip those steps if paths are missing.

**Beamtime path:** After `list_beamtimes`, set `BEAMTIME` to a directory path equal to a value in the `beamtime_path` column (canonical path from the machine that ran ingest).

```python
import os
from pathlib import Path

import polars as pl
from IPython.display import display

from pyref.io import list_beamtimes, naming_qc_with_db_parse_flags, read_beamtime
from pyref.io.catalog_path import resolve_catalog_path

env_db = os.environ.get("PYREF_CATALOG_DB")
CATALOG = Path(env_db).expanduser().resolve() if env_db else resolve_catalog_path()
print("catalog.db:", CATALOG)
```

```python
bts = list_beamtimes(CATALOG)
bts
```

```python
# Set to one path from bts['beamtime_path'].
BEAMTIME = Path("/path/from/list_beamtimes/column")
view = read_beamtime(BEAMTIME, catalog_path=CATALOG, ingest=False)
print("samples:", view.entries.samples)
print("frame rows:", view.frames.height)
```

```python
qc = naming_qc_with_db_parse_flags(view.frames, CATALOG)
for name, df in qc.items():
    print(name, df.height)
    if df.height > 0:
        display(df)
```
