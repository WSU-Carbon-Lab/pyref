```python
df = pd.DataFrame({
    "A": [1,1.5,.5, 4, 5, 6],
    "B": [4,4.5,3.5, 7, 8, 9],

})

df
```

```python
def addAverageCol(df, averagePoints: int):
    average = df.iloc[:averagePoints].mean(axis=0).to_frame().T
    average.index = [averagePoints-1]
    df_pre = df.iloc[:averagePoints]
    df_pre.index = [0]*averagePoints
    df_post = df.iloc[averagePoints:]
    new_df = pd.concat([df_pre, average, df_post])
    return new_df
```

```python
addAverageCol(df, 3)
```
